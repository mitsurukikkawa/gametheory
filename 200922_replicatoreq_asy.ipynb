```
import numpy as np
from scipy.integrate import odeint
from matplotlib import pyplot as plt

#2-population replicator dynamic for 2x2 asymmetric game

# payoffs
a1=-4.0; a2=4.0; b1=4.0; b2=-4.0

#           A        B
#      A   a1, b1   0, 0
#      B   0,  0    a2, b2

#conflict of interest instead of common interest
def rep(pop_vec, t):
    s1, r1 = pop_vec
    return [ s1*(1-s1) * (a1*r1 - a2*(1-r1)) , r1*(1-r1)*(b1*s1 - b2*(1-r1)) ]

t = np.arange(0, 100, .005)

#y0 = [.3, .4]
#y = odeint(rep, y0, t)
#plt.plot(y[:,1], y[:,0]) 

for s0 in np.arange(0.05, 0.95, .1):
    for r0 in np.arange(0.05, 0.95, .5):
        y0 = [s0, r0]
        y = odeint(rep, y0, t)
        plt.plot(y[:,1], y[:,0]) 

plt.show()
```
